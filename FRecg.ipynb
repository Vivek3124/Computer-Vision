```python
known_names = []
images_path = "/content/drive/MyDrive/cfp-dataset/Images/"
names_path = "/content/sample_data/list_name.txt"
file1 = open(names_path, 'r')
while True:
    # Get next line from file
    line = file1.readline()
    # if line is empty
    # end of file is reached
    if not line:
        break
    known_names.append(line[:-1])

known_names
```

```python
images_no_list = os.listdir(images_path)
images_no_list.sort()
```

```python
if not fr.face_encodings(cv2.imread(images_path + "036/frontal/07.jpg")):
  print(True)
# cv2_imshow(cv2.imread(images_path + "036/frontal/07.jpg"))
```

```python
known_name_encodings = []
for image_no in images_no_list:
  encodings = []
  image_path = images_path + image_no + "/frontal/"
  images = os.listdir(image_path)
  images.sort()
  print(images[:-2])
  for img in images[:-2]:
    im = fr.load_image_file(image_path + img)
    temp = fr.face_encodings(im)
    if not temp:
      continue
    print("\tEncoding image "+img+" of "+image_no)
    encoding = temp[0]
    encodings.append(encoding)
  known_name_encodings.append(encodings)
  # known_names.append(os.path.splitext(os.path.basename(image_path))[0].capitalize())
```

```python
test_images = os.listdir("/content/test_images/")
test_images
```

```python
test_images[5][0]
```

```python
results = []
for test_image in test_images:
  if test_image[0] == '.':
    continue
  image = fr.load_image_file("/content/test_images/"+test_image)
  face_encod = fr.face_encodings(image)[0]
  face_distances_list = []
  name = ""
  for known_name_encoding in known_name_encodings:
    face_distances = fr.face_distance(known_name_encoding, face_encod)
    best_match_index = np.argmin(face_distances)
    face_distances_list.append(face_distances[best_match_index])
    
  best_match = np.argmin(face_distances_list)
  name = known_names[best_match]
  results.append(name)
```
